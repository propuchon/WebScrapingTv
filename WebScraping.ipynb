# Web Scrapping

## Ticker

```python
URL = 'https://www.tradingview.com/symbols/SET-SET50/components/'
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")

# define tag and class we interested.
tag_ticker = soup.findAll("a",{"class":"apply-common-tooltip tickerNameBox-hMpTPJiS tickerName-hMpTPJiS"})
tag_ticker
```

```python
ticker_list = []

for i in tag_ticker:
    ticker_list.append(i.text)

ticker_list
```

## Market Cap

```python
URL = 'https://www.tradingview.com/symbols/SET-SET50/components/'
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")

# 9 rows per 1 tickers (format)
tag_market_info = soup.findAll("td",{"class":"cell-TKkxf89L right-TKkxf89L"})
tag_market_info
```

```python
market_info = []

for i in tag_market_info:
    market_info.append(i.text)
market_info
```

```python
# list[start::step]
market_cap_list = market_info[5::9]

# ref: https://bobbyhadz.com/blog/python-get-every-nth-element-of-list
```

# Data Preparation

preparation data for write into `pine script`

```python
df = pd.DataFrame({'ticker_name': ticker_list, 'market_cap': market_cap_list})
```

```python
df #format number to order the list
```

```python
df.info()
```

```python
# data prep. to 'market_cap' for sorting
df['numbers'] = df['market_cap'].str[:-4]
df['numbers'] = df['numbers'].astype(float)
df['unit'] = df['market_cap'].str[-4:]
df['unit_number'] = [1 if i == 'TTHB' else 0 for i in df['unit']]

# sort by market cap for each ticker
df.sort_values(['unit_number', 'numbers'], ascending=False, inplace=True)
df.reset_index(inplace=True, drop=True)
df
```

```python
# format for pine script
# sec_1 = request.security('AOT', timeframe.period, output)
# Limitation request.*() — A script cannot make more than 40 calls 
for i in range(0, 40):
    print(f"sec_{i+1} = request.security('{df['ticker_name'][i]}', timeframe.period, output)")
```

```python
import talib as ta
from tvDatafeed import TvDatafeed, Interval


tv = TvDatafeed()
```

```python
aot_ticker = tv.get_hist(symbol='AOT', exchange='SET', interval=Interval.in_daily, n_bars=5000)
gulf_ticker = tv.get_hist(symbol='GULF', exchange='SET', interval=Interval.in_daily, n_bars=5000)
```

```python
aot_ticker['sma_20'] = ta.SMA(aot_ticker['close'], timeperiod=20)
gulf_ticker['sma_20'] = ta.SMA(gulf_ticker['close'], timeperiod=20)
```

```python
# close > sma = uptrend (monthly)
aot_ticker['trend'] = aot_ticker['close'] > aot_ticker['sma_20'] 
gulf_ticker['trend'] = gulf_ticker['close'] > gulf_ticker['sma_20']
```
